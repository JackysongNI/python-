```python
import cv2
import numpy as np 

image = cv2.imread("images/input.jpg")

# 創建一個增強矩陣
M = np.ones(image.shape, dtype="uint8") * 75

added = cv2.add(image, M) # 相加
cv2.imshow("Added",added) # 圖片變亮

subtracted = cv2.subtract(image,M) # 相減
cv2.imshow("subtracted", subtracted)

cv2.waitKey()
cv2.destroyAllWindows()
```
