```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('./images/input.jpg')
histogram = cv2.calcHist([image],[0],None,[256],[0,256])

# cv2.calcHist([图像]，[通道]，掩码，［范围］,[BIN的数量表示尺寸])
# calcHist 帮助我们绘图（直方图）
# 图像；uint8 或者，uint32的图像要用
# 通道：对于彩色图像来说可以分别使用［0］［1］［2］的方式分别来计算 蓝色 绿色 或是红色
# 掩码：掩盖掉图片
# hitsize：表示图表的单位位置

# 整张图片里面蓝色的分布情况
# pit.hist(image.ravel(),256,[0,256]);plt.show()

color = ("b","g","r")
for i, col in enumerate(color):
    histogram2 = cv2.calcHist([image],[i],None,[256],[0,256])
    plt.plot(histogram2, color = col)
    plt.xlim([0,256])
plt.show()
```
